# Limpando a Sujeira e Salvando o Ouro

Este notebook serve para limpar os dados dos arquivos de fornecidos pelo governo de São Paulo sobre a remuneração dos Serviços Públicos do estado.

## 1. Importações

```python
import numpy as np # Algebra linear
import pandas as pd # Manipulação de dados
import pprint # Impressão formatada

import os 
paths = {}
month_12 = []
month_13 = []
month_14 = []
month_15 = []
month_16 = []
month_17 = []
month_18 = []
month_19 = []
month_20 = []
for dirname, _, filenames in os.walk('../input/'):
    for filename in filenames:
        path = os.path.join(dirname, filename)
        
        if path.find('.txt') != -1:
            year = dirname[-4:]
            if year == '2020':
                month_20.append(path)
            elif year == '2019':
                month_19.append(path)
            elif year == '2018':
                month_18.append(path)
            elif year == '2017':
                month_17.append(path)
            elif year == '2016':
                month_16.append(path)
            elif year == '2015':
                month_15.append(path)
            elif year == '2014':
                month_14.append(path)
            elif year == '2013':
                month_13.append(path)
            elif year == '2012':
                month_12.append(path)
            
paths['2020'] = month_20   
zero = month_20 

paths['2019'] = month_19    
nove = month_19

paths['2018'] = month_18
oito = month_18

paths['2017'] = month_17   
sete = month_17

paths['2016'] = month_16
seis_1 = month_16[6:]
seis_2 = month_16[:6]

paths['2015'] = month_15  
cinco_1 = month_15[6:]
cinco_2 = month_15[:6]

paths['2014'] = month_14
quatro_1 = month_14[6:]
quatro_2 = month_14[:6]

paths['2013'] = month_13   
tres_1 = month_13[6:]
tres_2 = month_13[:6]   

paths['2012'] = month_12
dois_1 = month_12

pprint.pprint(paths)
```

```python
for path in dois_1:
    df = pd.read_csv(path, encoding='iso-8859-1', sep=';', index_col=0)

    try:
        df.columns = ['Name', 'Job', 'Organ', 'Total_remuneration', 
                      'Goss_salary', 'Net_salary']
        numerical = ['Total_remuneration', 'Goss_salary', 
                     'Net_salary']
    finally:

        for column in numerical:
            df[column] = df[column].str.replace(',', '.').astype('float32')

        df.fillna('None', inplace=True)
        
        df.to_csv(f"{path[:-4]}.csv")
        print(f'OK: {path[:-4]}.csv')
```
